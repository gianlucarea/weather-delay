```python
# Importing the libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Normalization
#from sklearn.metrics import r2_score
#from tensorflow_addons.metrics import RSquare
from keras import backend as K
```

```python
# Import dataset
main_dataset = pd.read_csv("../processed_files/main_dataset.csv", index_col=[0])


```

```python
main_dataset['ritardo'] = main_dataset.ritardo_partenza + main_dataset.ritardo_arrivo
```

```python
main_dataset = main_dataset.drop(['trip_id','codice_stazione','codice',
                                  'data_giorno','stop_lat','stop_lon','orario_arrivo','orario_partenza',
                                  'orario','arrival_time','departure_time','stop_sequence',
                                  'direction_id','stop_code','time','ritardo'
                                 ], axis=1)

main_dataset = main_dataset.round({'temperature':0,'speed':0,'wind_speed':0,'temperature_linea':0,
                                  'app_temp_linea':0,'wind_dir_10m_linea':0,'soil_temperature_linea':0 ,
                                  'wind_gusts_linea':0, 'wind_dir_100m_linea':0 ,'wind_speed_100m_linea':0,
                                  'wind_speed_10m_linea':0, 'wmo_code_linea':0 , 'humidity_linea':0}) 
```

```python
main_dataset
```

```python
main_dataset = main_dataset.astype(np.float64)
main_dataset=(main_dataset-main_dataset.min())/(main_dataset.max()-main_dataset.min())
#main_dataset.ritardo_partenza=(main_dataset.ritardo_partenza-main_dataset.ritardo_partenza.min())/(main_dataset.ritardo_partenza.max()-main_dataset.ritardo_partenza.min())
```

```python
# Constructing the input
temp = main_dataset.copy()
temp = temp.drop('ritardo_arrivo',axis=1)
X = temp.values
Y = main_dataset.ritardo_arrivo.values

X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X, Y, test_size=0.3,random_state=42)
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5,random_state=42)
```

```python
# Defining the model
model = keras.Sequential([
    keras.layers.Dense(17, input_shape=(34,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.AlphaDropout(0.1),
    keras.layers.Dense(32,  activation='relu'),
    keras.layers.AlphaDropout(0.1),
    keras.layers.Dense(16,  activation='relu'),
    keras.layers.AlphaDropout(0.1),
    keras.layers.Dense(8,  activation='relu'),
    keras.layers.AlphaDropout(0.1),
#    keras.layers.Dense(64, kernel_initializer='lecun_normal', activation='selu'),
#    keras.layers.AlphaDropout(0.1),
#    keras.layers.Dense(32, kernel_initializer='lecun_normal', activation='selu'),
#    keras.layers.AlphaDropout(0.1),
#    keras.layers.Dense(16, kernel_initializer='lecun_normal', activation='selu'),
#    keras.layers.AlphaDropout(0.1),
#    keras.layers.Dense(8, kernel_initializer='lecun_normal', activation='selu'),
#    keras.layers.AlphaDropout(0.1),
    keras.layers.Dense(1, activation='linear')
])
 
optimizer = keras.optimizers.Adam(lr=0.01)
model.compile(loss='mse',
              optimizer=optimizer,
              metrics=['mse',tf.keras.metrics.RootMeanSquaredError(),'mae'])


# fitting the model
model.fit(X_train, Y_train, epochs=100, batch_size=4,  validation_data=(X_val, Y_val))
```

```python
model.evaluate(X_test, Y_test, batch_size=16)
```
