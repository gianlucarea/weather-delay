### Clean up time in measurment table

```python
measurments = measurments.assign(data_giorno=None)
measurments = measurments.assign(orario=None)
measurments = measurments.assign(wmo_code=None)
for index, row  in measurments.iterrows():
    if(index % 10000 == 0):
        print("row " + str(index) + "/" +str(measurments.index.max()))
    date,time = measurments.loc[index,"time"].split("T")
    measurments.loc[index,'orario'] = time
    measurments.loc[index,'data_giorno'] = date.replace("-","")
```

```python
for index, row  in wmo_codes.iterrows():
    if(index % 1000 == 0):
        print("row " + str(index) + "/" +str(wmo_codes.index.max()))
    wmo_codes.loc[index,'time'] = wmo_codes.loc[index,"time"].replace("-","")
```

```python
measurments.to_csv("../../weather/measurments_after_split_clock.csv")
wmo_codes.to_csv("../../weather/wmo_codes_after_split.date.csv")
```
